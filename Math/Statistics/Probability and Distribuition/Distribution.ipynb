```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
random_vector = np.random.randint(1, 11, size=10000)
```

```python
# Create a histogram
plt.hist(random_vector,color='blue', alpha=0.7, edgecolor='black')

# Add titles and labels
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
