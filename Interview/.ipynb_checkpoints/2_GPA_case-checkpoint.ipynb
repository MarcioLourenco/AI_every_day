## Internalização do "Meu Desconto" no GPA (Pão de Açúcar)

Na época em que trabalhava no GPA (Pão de Açúcar), participei de um projeto estratégico de internalização da solução "Meu Desconto", que até então era gerenciada por uma consultoria externa. Esse processo foi essencial para aumentar o controle sobre o algoritmo de recomendação de produtos por cliente, que era considerado uma "caixa preta" devido à falta de transparência nos métodos utilizados pela consultoria.

#### Desenvolvimento do Algoritmo Interno:

Criamos um algoritmo de machine learning focado em recomendações personalizadas de produtos para clientes.
Após várias iterações e otimizações, conseguimos alcançar uma acurácia de 80%, superando nossas expectativas iniciais e aproximando o desempenho ao da solução terceirizada.
- **Processamento em Batch**: Implementamos um processo de execução em batches, realizado **quinzenalmente**, para garantir que as recomendações estivessem sempre atualizadas com base nos dados mais recentes dos clientes e no histórico de compras.
- **Arquitetura da Solução**: Utilizamos **Airflow** como orquestrador, que acionava o **AWS Data Pipeline** para gerenciar a execução de jobs Spark em um cluster **Amazon EMR**.
O cluster processava as recomendações em larga escala e os resultados eram gravados em um banco de dados **MongoDB**, permitindo consultas rápidas e acessos eficientes.
